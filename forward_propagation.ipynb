we know,<br>
=>&ensp; $z = w_1x_1+ w_2x_2+ ... + w_nx_n+ b$
<p> in matrix form, 

$\text{W = } \begin{bmatrix}
w_1 \\
w_2 \\
. \\
. \\
. \\
w_n
\end{bmatrix}
,
\quad
\text{X = } \begin{bmatrix}
x_1\\
x_2\\
. \\
. \\
. \\
x_n
\end{bmatrix}
,
\quad
\text{b = } \begin{bmatrix}
b\\
\end{bmatrix}
$

=>&ensp; 
$z = \text{W}^T\text{X}+ \text{b}$ 
<p>

where, $\text{W}$ is weight matrix, $\text{X}$ is input matrix and  $\text{b}$ is bias matrix.

#### (i) **For Layer #1** 

</font>


$\text{W}_1 = 
\begin{bmatrix}
w_{11}^1 & w_{12}^1 & w_{13}^1\\
\\
w_{21}^1 & w_{22}^1 & w_{23}^1\\
\\
w_{31}^1 & w_{32}^1 & w_{33}^1\\
\\
w_{41}^1 & w_{42}^1 & w_{43}^1
\end{bmatrix}
,
\quad
\text{X}_1 = 
\begin{bmatrix}
x_{i1} \\
\\
x_{i2} \\
\\
x_{i3} \\
\\
x_{i4} \\
\end{bmatrix}
,
\quad
\text{b}_1 = 
\begin{bmatrix}
b_{11}\\
\\
b_{12}\\
\\
b_{13}
\end{bmatrix}
$

<p>

We know, <p>

=> $z_1 = \text{W}_1^T\text{X}_1+ \text{b}_1 \quad$ and 
 $\quad \text{W}_1^T = \begin{bmatrix}
w_{11}^1 & w_{21}^1 & w_{31}^1 & w_{41}^1 \\
w_{12}^1 & w_{22}^1 & w_{32}^1 & w_{42}^1 \\
w_{13}^1 & w_{23}^1 & w_{33}^1 & w_{43}^1
\end{bmatrix}$

=> $z_1 = 
\begin{bmatrix}
(w_{11}^1 x_{i1} + w_{21}^1 x_{i2} + w_{31}^1 x_{i3} + w_{41}^1 x_{i4}) + b_{11} \\
\\
(w_{12}^1 x_{i1} + w_{22}^1 x_{i2} + w_{32}^1 x_{i3} + w_{42}^1 x_{i4}) + b_{12} \\
\\
(w_{13}^1 x_{i1} + w_{23}^1 x_{i2} + w_{33}^1 x_{i3} + w_{43}^1 x_{i4}) + b_{13}
\end{bmatrix}
$

<p>

=> $\sigma({z_1}) = \sigma \big(
\begin{bmatrix}
(w_{11}^1 x_{i1} + w_{21}^1 x_{i2} + w_{31}^1 x_{i3} + w_{41}^1 x_{i4}) + b_{11} \\
\\
(w_{12}^1 x_{i1} + w_{22}^1 x_{i2} + w_{32}^1 x_{i3} + w_{42}^1 x_{i4}) + b_{12} \\
\\
(w_{13}^1 x_{i1} + w_{23}^1 x_{i2} + w_{33}^1 x_{i3} + w_{43}^1 x_{i4}) + b_{13}
\end{bmatrix} \big) = \begin{bmatrix}O_{11}\\O_{12}\\O_{13}\end{bmatrix} = \text{a}^\text{[1]}
$


<font color='teal'> 

#### (ii) **For Layer #2** 

</font>

$\text{W}_2 = 
\begin{bmatrix}
w_{11}^2 & w_{12}^2\\
\\
w_{21}^2 & w_{22}^2\\
\\
w_{31}^2 & w_{32}^2
\end{bmatrix}
,
\quad
\text{X}_2 = 
\begin{bmatrix}O_{11}\\O_{12}\\O_{13}\end{bmatrix}
,
\quad
\text{b}_2 = 
\begin{bmatrix}
b_{21}\\
\\
b_{22}
\end{bmatrix}
$

<p>
Again, <p>

=> $z_2 = \text{W}_2^T\text{X}_2 + \text{b}_2 \quad$ and 
$\quad \text{W}_2^T = \begin{bmatrix}
w_{11}^2 & w_{21}^2 & w_{31}^2 \\
w_{12}^2 & w_{22}^2 & w_{32}^2
\end{bmatrix}
$
<p>

=> $\sigma({z_2}) = \sigma \big(
\begin{bmatrix}
(w_{11}^2 O_{11} + w_{21}^2 O_{12} + w_{31}^2 O_{13}) + b_{21} \\
(w_{12}^2 O_{11} + w_{22}^2 O_{12} + w_{32}^2 O_{13}) + b_{22}
\end{bmatrix} \big) = \begin{bmatrix}O_{21}\\O_{22}\end{bmatrix} = \text{a}^\text{[2]}
$

<font color='teal'> 

#### (iii) **For Layer #3** 

</font>

$\text{W}_3 = 
\begin{bmatrix}
w_{11}^3\\
\\
w_{21}^3
\end{bmatrix}
,
\quad
\text{X}_3 = 
\begin{bmatrix}O_{21}\\O_{22}\end{bmatrix}
,
\quad
\text{b}_3 = 
\begin{bmatrix}
b_{31}\\
\end{bmatrix}
$

<p>
Again, <p>

=> $z_3 = \text{W}_3^T\text{X}_3 + \text{b}_3 \quad$ and 
$\quad \text{W}_3^T = \begin{bmatrix}
w_{11}^3 & w_{21}^3
\end{bmatrix}
$
<p>

=> $\sigma({z_3}) = \sigma \big(
\begin{bmatrix}
(w_{11}^3 O_{21} + w_{21}^3 O_{22}) + b_{31}
\end{bmatrix} \big) = \begin{bmatrix}O_{31}\end{bmatrix} = \text{a}^\text{[3]} = \hat{y} \quad\text{or say, Final Output}
$
